```python
import os
import matplotlib.pyplot as plt
import librosa, librosa.display
import IPython.display as ipd
```

```python
violin_sound_file = "/content/audio_resources_violin_c.wav"
piano_sound_file = "/content/audio_resources_piano_c.wav"
tremolo_sound_file = "/content/audio_resources_tremolo.wav"
```

```python
# loading the sound files
violin_sd, violin_sr = librosa.load(violin_sound_file, sr=None)
piano_sd, piano_sr = librosa.load(piano_sound_file, sr=None)
tremolo_sd, tremolo_sr = librosa.load(tremolo_sound_file, sr=None)
```

```python
# Checking out the sample rates of the sound files
print(violin_sr, piano_sr, tremolo_sr)
```

```python
## Plotting the spectogram for each audio file
def plot_spectrogram(signal, name):
  """ Compute power spectrogram with Short-Term Fourier Transforms and plot result"""
  spectrogram = librosa.amplitude_to_db(librosa.stft(signal))
  plt.figure(figsize=(18,15))
  librosa.display.specshow(spectrogram, y_axis='log')
  plt.colorbar(format="%+2.0f dB")
  plt.title(f"Log-frequency power spectrogram for {name}")
  plt.xlabel("Time")
  plt.show()
```

```python
# Checking out the violin audio and its spectrogram
ipd.Audio(violin_sd, rate=violin_sr)
```

```python
plot_spectrogram(violin_sd, "Violin")
```

```python
# Checking out the piano audio and its spectrogram
ipd.Audio(piano_sd, rate=piano_sr)
```

```python
plot_spectrogram(piano_sd, "Piano")
```

```python
# Checking out the tremolo audio and its spectrogram
ipd.Audio(tremolo_sd, rate=tremolo_sr)
```

```python
plot_spectrogram(tremolo_sd, "Tremolo")
```
